```python
import json
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Uploading to Data Frame
file_path = '/content/skullcandyproduct.json' #Used for respective file
with open(file_path, 'r') as file:
    data = json.load(file)

reviews_data = []
count = 0
for result in data['results']:
  asin = result['content']["asin"]
  for review in result['content']['reviews']:
    count += 1
    try:
      review_info = {
      'review_id': review['id'],
      'product_id': asin,
      "title": review["title"],
      "author": review["author"],
      "rating": review["rating"],
      "content": review["content"],
      "timestamp": review["timestamp"],
      "profile_id": review["profile_id"],
      "is_verified": review["is_verified"],
      "helpful_count": review.get('helpful_count', 0),
      "product_attributes": review["product_attributes"]
      }
      reviews_data.append(review_info)
    except:
      print(count)
      break

reviews_df = pd.DataFrame(reviews_data)
reviews_df.to_csv('reviews.csv', index=False)
reviews_df
```

```python
#Merging CSV files
df_1 = pd.read_csv('/content/applereviews.csv')
df_2 = pd.read_csv('/content/beatsreview.csv')
df_3 = pd.read_csv('/content/beatsreviews2.csv')
df_4 = pd.read_csv('/content/bosereviews.csv')
df_5 = pd.read_csv('/content/jabrareviews.csv')
df_6 = pd.read_csv('/content/jblreviews.csv')
df_7 = pd.read_csv('/content/sennheiserreviews.csv')
df_8 = pd.read_csv('/content/skullcandyreviews.csv')
df_9 = pd.read_csv('/content/sonyreviews.csv')
df_10 = pd.read_csv('/content/soundcorereviews.csv')

merged_df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10], ignore_index=True)
merged_df
merged_df.to_csv('mergedreviews.csv', index=False)
merged_df
```
